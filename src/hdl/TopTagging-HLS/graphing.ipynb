```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- GRU: Handmade vs HLS full-bitwidth sweep comparison ---

gru_resf, gru_res = plt.subplots()
gru_timingf, gru_timing = plt.subplots()
gru_accf, gru_acc = plt.subplots()
gru_latf, gru_lat = plt.subplots()

gru_colors = ['tab:blue', 'tab:orange']
gru_marker_styles = ['o', 's']
gru_line_styles = ['-', '--']
gru_legend_lines = []
gru_color_idx = [0]

renames_gru = {"Slice Registers": "Registers", "Slice LUTs": "LUTs"}

# Columns to skip in the resource plot
RES_SKIP = {"Bits", "Accuracy", "Timing", "Bonded IOB",
            "Block RAM Tile", "Cycle Latency", "Total Latency", "II", "Total II"}

def plot_res_gru(file, ls, name, marker):
    df = pd.read_csv(file, delimiter=", ", engine="python")
    x = df["Bits"]
    c = gru_colors[gru_color_idx[0] % len(gru_colors)]
    mk = marker

    # Timing (clock period = 10 - timing slack proxy)
    gru_timing.plot(x, df["Timing"], label=name, linestyle=ls, color=c, marker=mk)

    # Accuracy
    gru_acc.plot(x, (df["Accuracy"]) * 100, label=name, linestyle=ls, color=c, marker=mk)

    # Total Latency (only present in HLS CSV)
    if "Total Latency" in df.columns:
        gru_lat.plot(x, df["Total Latency"], label=name, linestyle=ls, color=c, marker=mk)
    else:
        gru_lat.plot(x, 141 * df["Timing"], label=name, linestyle=ls, color=c, marker=mk)

    # Resources — skip non-resource columns
    res_cols = [k for k in df.columns if k.strip() not in RES_SKIP]
    for i, key in enumerate(res_cols):
        label = renames_gru.get(key.strip(), key.strip())
        gru_res.plot(x, df[key],
                     label=label if gru_color_idx[0] == 0 else None,
                     linestyle=ls, color=plt.cm.tab10(i / 10), marker=mk)

    gru_legend_lines.append((name, ls, mk, c))
    gru_color_idx[0] += 1


plot_res_gru("GRU-Handmade/Results/util_handmade_toptag_gru_fullsweep_fixed.csv",
             '-', "Handmade GRU", 'o')
plot_res_gru("GRU-HLS/results/util_hls_toptag_gru_fullsweep.csv",
             '--', "HLS GRU", 's')

# --- Style: Resource ---
first_leg = gru_res.legend(loc='upper left')
gru_res.add_artist(first_leg)
impl_lines = [gru_res.plot([], [], label=n, linestyle=ls, marker=mk, color=c)[0]
              for n, ls, mk, c in gru_legend_lines]
gru_res.legend(handles=impl_lines, loc='upper right')
gru_res.set_title("GRU Resource Utilization: Handmade vs HLS")
gru_res.set_xlabel("Bitwidth")
gru_res.set_ylabel("Utilization (%)")
gru_res.set_xticks(np.arange(4, 40, 3))
gru_res.set_xlim((4, 37))

# --- Style: Timing ---
gru_timing.legend(loc='upper left')
gru_timing.set_title("GRU Clock Period: Handmade vs HLS")
gru_timing.set_xlabel("Bitwidth")
gru_timing.set_ylabel("Clock Period Slack (ns)")
gru_timing.set_xticks(np.arange(4, 40, 3))
gru_timing.set_xlim((4, 37))

# --- Style: Accuracy ---
gru_acc.legend(loc='upper right')
gru_acc.set_title("Accuracy: Handmade vs HLS")
gru_acc.set_xlabel("Bitwidth")
gru_acc.set_ylabel("Accuracy (%)")
gru_acc.set_xticks(np.arange(4, 40, 3))
gru_acc.set_xlim((4, 37))

# --- Style: Latency ---
gru_lat.legend(loc='upper right')
gru_lat.set_title("GRU Total Latency: HLS")
gru_lat.set_xlabel("Bitwidth")
gru_lat.set_ylabel("Total Latency (ns)")
gru_lat.set_xticks(np.arange(4, 40, 3))
gru_lat.set_xlim((4, 37))
gru_lat.set_ylim(0)

plt.show()
```
