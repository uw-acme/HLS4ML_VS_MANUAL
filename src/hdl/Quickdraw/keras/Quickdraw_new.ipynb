```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, GRU, Dense, Dropout, Activation
from tensorflow.keras.models import load_model

inputs = Input(shape = (100,3))

x = GRU(128, return_sequences=True)(inputs)
x = Dropout(0.5)(x)
x = Dense(256)(x)
x = Activation('relu')(x)
x = Dropout(0.5)(x)
x = Dense(128)(x)
x = Activation('relu')(x)
x = Dense(5)(x)
predictions = Activation('softmax')(x)
gru = Model(inputs=inputs, outputs=predictions)

gru.summary()
```
