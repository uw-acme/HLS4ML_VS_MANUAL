```python
import numpy as np

NPY_FILE = "X_test.npy"
TXT_FILE = "X_test_20_10.txt"

TOTAL_BITS = 20
FRAC_BITS = 10


def float_to_fixed_bin(x, total_bits=20, frac_bits=10):

    fixed_val = int(round(float(x) * (2 ** frac_bits)))

    max_val = (1 << (total_bits - 1)) - 1
    min_val = -(1 << (total_bits - 1))

    fixed_val = max(min(fixed_val, max_val), min_val)

    if fixed_val < 0:
        fixed_val = (1 << total_bits) + fixed_val

    return format(fixed_val, f'0{total_bits}b')


# Load object array
data = np.load(NPY_FILE, allow_pickle=True)

# Take first 100 timesteps
data = data[:100]

with open(TXT_FILE, "w") as f:

    for timestep in data:

        # Convert timestep object into normal ndarray
        timestep = np.asarray(timestep).reshape(-1)

        for feature in timestep:

            bin_str = float_to_fixed_bin(feature)

            f.write(bin_str + "\n")
```
